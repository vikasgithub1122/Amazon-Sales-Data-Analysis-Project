```python
# Perform the groupby operation
profit_by_region = df.groupby('Region')['Total Profit'].sum()

# Create a bar plot
plt.figure(figsize=(10, 5))
profit_by_region.plot(kind='bar')

# Customize the plot
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add value labels on top of each bar
for i, v in enumerate(profit_by_region):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# Group by Region and calculate mean Total Profit
Avg_Profit_by_region = df.groupby('Region')['Total Profit'].mean()

# Set the Seaborn style
sns.set_style("whitegrid")

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 5))

# Create the pie chart
wedges, texts, autotexts = ax.pie(Avg_Profit_by_region,
                                  labels=Avg_Profit_by_region.index,
                                  autopct='%1.1f%%',
                                  startangle=90,
                                  wedgeprops=dict(width=0.5))  # This creates a donut chart

# Customize the plot
plt.title('Average Total Profit Distribution by Region', fontsize=16)

# Add a legend
ax.legend(wedges, Avg_Profit_by_region.index,
          title="Regions",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

# Customize text properties
plt.setp(autotexts, size=10, weight="bold")
plt.setp(texts, size=12)

# Use Seaborn's color palette
colors = sns.color_palette("pastel")
for i, wedge in enumerate(wedges):
    wedge.set_facecolor(colors[i % len(colors)])

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# Assuming units_sold_by_year_month is already calculated
units_sold_by_year_month = df.groupby(['Year', 'Month'])['Units Sold'].sum().reset_index()

# Create a datetime column for proper ordering
units_sold_by_year_month['Date'] = pd.to_datetime(units_sold_by_year_month['Year'].astype(str) + '-' + units_sold_by_year_month['Month'].astype(str) + '-01')

# Sort by date
units_sold_by_year_month = units_sold_by_year_month.sort_values('Date')

# Create a string representation of Year-Month for x-axis labels
units_sold_by_year_month['Year-Month'] = units_sold_by_year_month['Date'].dt.strftime('%Y-%m')

# Set up the plot
plt.figure(figsize=(15, 6))
sns.set_style("whitegrid")

# Create the bar chart
ax = sns.barplot(x='Year-Month', y='Units Sold', data=units_sold_by_year_month)

# Customize the plot
plt.title('Total Units Sold by Year and Month', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)

# Rotate x-axis labels
plt.xticks(rotation=90)

# Show only every nth label to prevent overcrowding
n = len(units_sold_by_year_month) // 10  # Adjust this value to show more or fewer labels
for idx, label in enumerate(ax.get_xticklabels()):
    if idx % n != 0:
        label.set_visible(False)

# Add value labels on top of each bar
for i, v in enumerate(units_sold_by_year_month['Units Sold']):
    ax.text(i, v, f'{v:,}', ha='center', va='bottom', fontsize=8)

# Use a tight layout
plt.tight_layout()

# Show the plot
plt.show()
```
